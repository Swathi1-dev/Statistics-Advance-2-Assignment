#Theory

1. What is hypothesis testing in statistics?

Ans: A claim / assumption which has been made and testing it using statistic methods.

2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans: The null hypothesis (H0) says that there is no significant relationship or difference between variables, while the alternative hypothesis (Ha) suggests that there is a significant relationship or difference

3.  What is the significance level in hypothesis testing, and why is it important?

Ans:In hypothesis testing, the significance level, denoted by Î± (alpha), is the probability of rejecting the null hypothesis when it is actually true.

4.  What does a P-value represent in hypothesis testing?

Ans: P value is used to reject or fail to reject the null hypothesis.

5.  How do you interpret the P-value in hypothesis testing?

Ans: In hypothesis testing, the p-value is interpreted by comparing it to a predetermined significance level (alpha, usually 0.05).

6.  What are Type 1 and Type 2 errors in hypothesis testing?

Ans: In hypothesis testing, Type 1 and Type 2 errors are two types of incorrect conclusions that can be drawn when testing a null hypothesis.

7.  What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

Ans: In hypothesis testing, a one-tailed test checks for a difference in one specific direction (e.g., greater than or less than), while a two-tailed test checks for a difference in either direction (e.g., greater than or less than).

8.  What is the Z-test, and when is it used in hypothesis testing?

Ans: A z-test is a statistical method used to test hypothesis about population parameters, particularly the population mean, when the population standard deviation is known or when the sample size is large enough (typically greater than 30) .

9.  How do you calculate the Z-score, and what does it represent in hypothesis testing?

Ans: A Z-score represents how many standard deviations a particular data point is from the mean of a dataset. It's calculated by subtracting the mean from the data point and then dividing by the standard deviation.

10.  What is the T-distribution, and when should it be used instead of the normal distribution?

Ans: The t-distribution is a probability distribution similar to the normal distribution, but with heavier tails, meaning it has a greater chance of extreme values. It's used when dealing with small sample sizes (typically less than 30) or when the population standard deviation is unknown

11.  What is the difference between a Z-test and a T-test?

Ans: The main difference between a Z-test and a T-test lies in the sample size and the knownness of the population standard deviation.

12.  What is the T-test, and how is it used in hypothesis testing?

Ans: A t-test is a statistical hypothesis test used to determine if there's a significant difference between the means of two groups.

13.  What is the relationship between Z-test and T-test in hypothesis testing?

Ans: Both z-tests and t-tests are used to determine if there's a significant difference between groups or if the sample mean significantly differs from a known population mean.

14.  What is a confidence interval, and how is it used to interpret statistical results?

Ans: A confidence interval is a range of values within which the true population parameter is likely to fall, based on a sample.

15.  What is the margin of error, and how does it affect the confidence interval?

Ans: The margin of error is a statistical measure that reflects the precision of an estimate, essentially how much the sample results might differ from the real population results

16.  How is Bayes' Theorem used in statistics, and what is its significance?

Ans: Bayes' Theorem is a mathematical formula that helps determine the conditional probability of an event, given that another event has already occurred.

17.  What is the Chi-square distribution, and when is it used?

Ans:The chi-square distribution is a continuous probability distribution often used in hypothesis testing, particularly with categorical data.

18.  What is the Chi-square goodness of fit test, and how is it applied?

Ans:The chi-square goodness-of-fit test is a statistical test used to determine if a sample data distribution matches a hypothesized population distribution.

19.  What is the F-distribution, and when is it used in hypothesis testing?

Ans:The F-distribution is a continuous probability distribution used in hypothesis testing, particularly when comparing variances or means between groups.

20.  What is an ANOVA test, and what are its assumptions?

Ans: An ANOVA (Analysis of Variance) test is a statistical method used to compare the means of two or more groups. It helps determine if there's a significant difference between group means by analyzing the variance within and between the groups. Key assumptions for ANOVA include: independent observations, normally distributed data, and equal variances (homoscedasticity).

```python
#6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#uniform distribution?


import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(a, b):

    if not isinstance(a, int) or not isinstance(b, int):
        raise ValueError("Both 'a' and 'b' must be integers.")
    if a > b:
        raise ValueError("'a' must be less than or equal to 'b'.")

    x = np.arange(a, b + 1)
    cdf = np.cumsum(np.ones_like(x)) / len(x)

    plt.figure(figsize=(8, 6))
    plt.step(x, cdf, where='post')
    plt.xlabel('Value')
    plt.ylabel('Cumulative Probability')
    plt.title('CDF of Discrete Uniform Distribution')
    plt.grid(True)
    plt.show()

# Example usage:
discrete_uniform_cdf(1, 6)


```

```python
#14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
#distribution.

import numpy as np
import scipy.stats as st

def calculate_z_score(data_point, mean, std_dev):

  z_score = (data_point - mean) / std_dev
  return z_score

def compare_to_standard_normal(z_score):

  probability_left = st.norm.cdf(z_score)
  probability_right = 1 - probability_left
  return probability_left, probability_right

if __name__ == "__main__":
    data_point = np.random.exponential(scale=2,size=100)
    mean = np.mean(data_point)
    std_dev = float(input("Enter the standard deviation of the dataset: "))

    z = calculate_z_score(data_point, mean, std_dev)
    print(f"The Z-score is: {z}")

    prob_left, prob_right = compare_to_standard_normal(z)
    print(f"Probability to the left of the Z-score: {prob_left}")
    print(f"Probability to the right of the Z-score: {prob_right}")
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.

# Import required libraries
from scipy import stats
import math

# Step 1: Define the known values
population_mean = 100       # Î¼
population_std_dev = 15     # Ï
sample_size = 45            # n
sample_mean = 105           # xÌ

# Step 2: Calculate Standard Error of the Mean (SEM)
standard_error = population_std_dev / math.sqrt(sample_size)

# Step 3: Calculate the z-score
z_score = (sample_mean - population_mean) / standard_error

# Step 4: Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Step 5: Print results
print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.4f}")

# Step 6: Significance level (Î±) for hypothesis testing
alpha = 0.05

# Step 7: Conclusion
if p_value < alpha:
    print("Reject the null hypothesis: There is enough evidence to suggest the average IQ has changed.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to suggest the average IQ has changed.")
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats  # For statistical functions

# Sample data: Salaries of 100 employees
np.random.seed(0)
salaries = np.random.normal(loc=55000, scale=5000, size=100)

# Point Estimate
point_estimate = np.mean(salaries)
print("Point Estimate (Mean Salary):", round(point_estimate, 2))

# 95% Confidence Interval
conf_level = 0.95
conf_interval = stats.t.interval(
    conf_level, # Confidence level
    df=len(salaries) - 1, # Degrees of freedom. It is the number of independent pieces of information
    loc=point_estimate, # Sample mean
    scale=stats.sem(salaries) # Standard error of the mean
)

print("95% Confidence Interval:", (round(conf_interval[0], 2), round(conf_interval[1], 2)))
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def z_test_two_tailed(sample_mean, mu_0, sigma, n, alpha=0.05):

    # Calculate the Z-score
    z_score = (sample_mean - mu_0) / (sigma / np.sqrt(n))

    # Calculate the critical Z-values for a two-tailed test
    z_critical = stats.norm.ppf(1 - alpha/2)

    # Decision: Reject null hypothesis if the Z-score is outside the critical region
    reject_null = abs(z_score) > z_critical

    # Plot the Z-test decision region
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, loc=0, scale=1)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue', lw=2)

    # Shading the rejection regions
    plt.fill_between(x, 0, y, where=(x <= -z_critical), color='red', alpha=0.5, label='Rejection Region (Left)')
    plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.5, label='Rejection Region (Right)')

    # Mark the critical values
    plt.axvline(-z_critical, color='red', linestyle='--', label=f'Critical Value: -{z_critical:.2f}')
    plt.axvline(z_critical, color='red', linestyle='--', label=f'Critical Value: {z_critical:.2f}')

    # Mark the calculated Z-score
    plt.axvline(z_score, color='green', linestyle='-', label=f'Z-score: {z_score:.2f}')

    plt.title('Two-Tailed Z-Test Decision Region')
    plt.xlabel('Z-Score')
    plt.ylabel('Probability Density')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()

    return z_score, reject_null

# Example usage of the function
sample_mean = 102  # Sample mean
mu_0 = 100        # Population mean under null hypothesis
sigma = 15         # Population standard deviation
n = 50            # Sample size
alpha = 0.05      # Significance level

z_score, reject_null = z_test_two_tailed(sample_mean, mu_0, sigma, n, alpha)

print(f"Calculated Z-score: {z_score:.2f}")
if reject_null:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testingE

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def hypothesis_test_error_visualization(mu_null, mu_alt, sigma, alpha=0.05, sample_size=30):

    # Step 1: Define the critical value (z-critical) based on alpha (Type 1 error rate)
    critical_value = stats.norm.ppf(1 - alpha, loc=mu_null, scale=sigma/np.sqrt(sample_size))

    # Step 2: Generate distributions for H0 and H1
    x = np.linspace(mu_null - 4 * sigma, mu_alt + 4 * sigma, 1000)
    dist_h0 = stats.norm.pdf(x, loc=mu_null, scale=sigma/np.sqrt(sample_size))
    dist_h1 = stats.norm.pdf(x, loc=mu_alt, scale=sigma/np.sqrt(sample_size))

    # Step 3: Calculate the Type 1 error (rejection of H0 when it is true)
    type_1_error_area = np.linspace(critical_value, mu_null + 4 * sigma, 1000)
    type_1_error = np.trapz(stats.norm.pdf(type_1_error_area, loc=mu_null, scale=sigma/np.sqrt(sample_size)), type_1_error_area)

    # Step 4: Calculate the Type 2 error (failure to reject H0 when H1 is true)
    type_2_error_area = np.linspace(mu_null - 4 * sigma, critical_value, 1000)
    type_2_error = np.trapz(stats.norm.pdf(type_2_error_area, loc=mu_alt, scale=sigma/np.sqrt(sample_size)), type_2_error_area)

    # Step 5: Plot the distributions and errors
    plt.figure(figsize=(10, 6))

    # Plot distributions
    plt.plot(x, dist_h0, label='H0: Null Hypothesis (mu_null)', color='blue', lw=2)
    plt.plot(x, dist_h1, label='H1: Alternative Hypothesis (mu_alt)', color='orange', lw=2)

    # Shade the Type 1 and Type 2 error areas
    plt.fill_between(x, 0, dist_h0, where=(x >= critical_value), color='red', alpha=0.5, label='Type 1 Error Area')
    plt.fill_between(x, 0, dist_h1, where=(x <= critical_value), color='green', alpha=0.5, label='Type 2 Error Area')

    # Labels and title
    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()

    # Return error rates
    return type_1_error, type_2_error

# Example usage of the function:
mu_null = 0  # Mean under null hypothesis
mu_alt = 2   # Mean under alternative hypothesis
sigma = 1    # Standard deviation of the population
alpha = 0.05 # Significance level (5%)
sample_size = 30  # Sample size

type_1_error, type_2_error = hypothesis_test_error_visualization(mu_null, mu_alt, sigma, alpha, sample_size)

print(f"Type 1 Error Rate: {type_1_error:.4f}")
print(f"Type 2 Error Rate: {type_2_error:.4f}")
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results.


import numpy as np
import scipy.stats as stats

def independent_t_test(sample1, sample2, alpha=0.05):
    """
    Perform an independent T-test and interpret the results.

    Parameters:
    sample1 (array-like): First sample data.
    sample2 (array-like): Second sample data.
    alpha (float): Significance level (default is 0.05).

    Returns:
    t_stat (float): The T-statistic.
    p_value (float): The p-value of the test.
    """
    # Perform the independent T-test
    t_stat, p_value = stats.ttest_ind(sample1, sample2)

    # Interpretation of the p-value
    if p_value < alpha:
        result = "Reject the null hypothesis: There is a significant difference between the means."
    else:
        result = "Fail to reject the null hypothesis: There is no significant difference between the means."

    # Return the results
    return t_stat, p_value, result

sample1 = np.array([23, 21, 18, 25, 30, 22, 20, 19, 24, 26])  # Sample 1
sample2 = np.array([33, 35, 31, 32, 37, 40, 38, 36, 39, 34])  # Sample 2

t_stat, p_value, result = independent_t_test(sample1, sample2, alpha=0.05)

# Display the results
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(result)
```

```python
#7. Perform a paired sample T-test using Python and visualize the comparison results.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def paired_t_test(sample1, sample2, alpha=0.05):

    t_stat, p_value = stats.ttest_rel(sample1, sample2)

    if p_value < alpha:
        result = "Reject the null hypothesis: There is a significant difference between the means."
    else:
        result = "Fail to reject the null hypothesis: There is no significant difference between the means."

    plt.figure(figsize=(10, 6))
    plt.plot(range(1, len(sample1) + 1), sample1, label='Before Treatment', marker='o', color='blue')
    plt.plot(range(1, len(sample2) + 1), sample2, label='After Treatment', marker='o', color='orange')

    plt.title('Paired Sample Comparison')
    plt.xlabel('Sample Index')
    plt.ylabel('Values')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()

    return t_stat, p_value, result


sample1 = np.array([23, 21, 18, 25, 30, 22, 20, 19, 24, 26])  # Before treatment
sample2 = np.array([28, 25, 20, 30, 35, 28, 24, 23, 30, 32])  # After treatment

# Perform the paired sample T-test
t_stat, p_value, result = paired_t_test(sample1, sample2, alpha=0.05)

# Display the results
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(result)
```

```python
#8. Simulate data and perform both Z-test and T-test, then compare the results using Python.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate data
np.random.seed(42)  # For reproducibility
sample_size = 30
population_mean = 100
population_std = 15

# Simulate sample data
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation

# Step 2: Perform Z-test
# Z-test is performed when we know the population standard deviation
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
# Calculate p-value for Z-test (two-tailed)
p_value_z = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Step 3: Perform T-test
# T-test uses the sample standard deviation
t_stat, p_value_t = stats.ttest_1samp(sample_data, population_mean)

# Step 4: Compare results
print("Z-test Results:")
print(f"Z-score: {z_score:.2f}")
print(f"Z-test p-value: {p_value_z:.4f}")

print("\nT-test Results:")
print(f"T-statistic: {t_stat:.2f}")
print(f"T-test p-value: {p_value_t:.4f}")

# Step 5: Interpretation
alpha = 0.05  # Significance level

# Z-test Interpretation
if p_value_z < alpha:
    z_result = "Reject the null hypothesis (Z-test): There is a significant difference."
else:
    z_result = "Fail to reject the null hypothesis (Z-test): No significant difference."

# T-test Interpretation
if p_value_t < alpha:
    t_result = "Reject the null hypothesis (T-test): There is a significant difference."
else:
    t_result = "Fail to reject the null hypothesis (T-test): No significant difference."

print("\nInterpretations:")
print(z_result)
print(t_result)

plt.figure(figsize=(10, 6))
plt.hist(sample_data, bins=10, alpha=0.7, label='Sample Data', color='blue')
plt.axvline(population_mean, color='red', linestyle='dashed', linewidth=2, label=f'Population Mean: {population_mean}')
plt.title('Histogram of Sample Data and Population Mean')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
    sample_data (array-like): The sample data for which we want to calculate the confidence interval.
    confidence_level (float): The desired confidence level (default is 0.95).

    Returns:
    lower_bound (float): The lower bound of the confidence interval.
    upper_bound (float): The upper bound of the confidence interval.
    """
    # Sample mean and standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation
    n = len(sample_data)  # Sample size

    # Calculate the critical value for the desired confidence level
    alpha = 1 - confidence_level
    z_critical = stats.norm.ppf(1 - alpha/2)  # For normal distribution

    # Calculate the margin of error
    margin_of_error = z_critical * (sample_std / np.sqrt(n))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
# Simulate some sample data (for example, exam scores)
np.random.seed(42)
sample_data = np.random.normal(loc=75, scale=10, size=100)  # Mean = 75, Std = 10, n = 100

# Calculate the 95% confidence interval
lower, upper = confidence_interval(sample_data, confidence_level=0.95)

print(f"95% Confidence Interval for the Sample Mean: ({lower:.2f}, {upper:.2f})")

```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
import scipy.stats as stats

def margin_of_error(sample_data, confidence_level=0.95, population_std=None):

    # Sample mean and sample size
    sample_mean = np.mean(sample_data)
    n = len(sample_data)  # Sample size

    # If population standard deviation is not provided, use the sample standard deviation
    if population_std is None:
        sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation
    else:
        sample_std = population_std

    # Calculate the critical value for the desired confidence level (Z or T)
    alpha = 1 - confidence_level
    z_critical = stats.norm.ppf(1 - alpha / 2)  # For normal distribution (Z-distribution)

    # Calculate the margin of error
    margin_of_error = z_critical * (sample_std / np.sqrt(n))

    return margin_of_error

np.random.seed(0)
sample_data = np.random.normal(loc=75, scale=10, size=100)

moe = margin_of_error(sample_data, confidence_level=0.95)

print(f"Margin of Error for 95% confidence level: {moe:.2f}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayesian_inference(prior, likelihood, evidence):

    # Bayes' Theorem: P(H|D) = (P(D|H) * P(H)) / P(D)
    posterior = (likelihood * prior) / evidence
    return posterior

prior = 0.01  # The prior probability that a person has the disease (1% of the population)
likelihood = 0.9  # The probability that the test is positive given the person has the disease (90% accuracy)
false_positive_rate = 0.05  # The probability that the test is positive given the person does NOT have the disease (5% false positive rate)

# P(D) = P(D|H)*P(H) + P(D|Â¬H)*P(Â¬H)
evidence = (likelihood * prior) + (false_positive_rate * (1 - prior))

# Calculate the posterior probability using Bayes' Theorem
posterior = bayesian_inference(prior, likelihood, evidence)

print(f"The probability that the person has the disease given a positive test result is: {posterior:.4f}")
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python.

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

data = {
    'Owns Car': [30, 10],   # 30 students own a car, 10 do not
    'Does Not Own Car': [5, 15]  # 5 students do not own a car, 15 do not
}

df = pd.DataFrame(data, index=['Student', 'Non-Student'])

print("Contingency Table:")
print(df)

# Perform Chi-Square test for independence
chi2, p_value, dof, expected = chi2_contingency(df)

print("\nChi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2}")
print(f"P-Value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies: \n{expected}")

# Interpretation of results
if p_value < 0.05:
    print("\nReject the null hypothesis: The two variables are dependent (there is an association).")
else:
    print("\nFail to reject the null hypothesis: The two variables are independent.")
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
#data.

import numpy as np

def calculate_expected_frequencies(observed_data):

    observed = np.array(observed_data)

    # Calculate the row totals, column totals, and grand total
    row_totals = observed.sum(axis=1)  # Sum along rows
    column_totals = observed.sum(axis=0)  # Sum along columns
    grand_total = observed.sum()  # Sum of all observed values

    # Calculate the expected frequencies using the formula
    expected = np.outer(row_totals, column_totals) / grand_total

    return expected

# Example observed data: Contingency table for two categorical variables
# For example, let's assume we have data on people who own cars and whether they are students or not.
observed_data = [
    [30, 10],  # 30 students own cars, 10 do not
    [5, 15]    # 5 non-students own cars, 15 do not
]

# Calculate the expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Display the observed and expected frequencies
print("Observed Frequencies:")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(expected_frequencies)
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (based on 60 rolls of a fair die)
observed = np.array([10, 12, 8, 11, 9, 10])

# Expected frequencies (for a fair die, each outcome should occur 1/6 of the time, i.e., 60/6 = 10 times)
expected = np.array([10, 10, 10, 10, 10, 10])

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(observed, expected)

# Output the results
print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Interpretation of results
if p_value < 0.05:
    print("\nReject the null hypothesis: The observed data do not follow the expected distribution.")
else:
    print("\nFail to reject the null hypothesis: The observed data follow the expected distribution.")
```
